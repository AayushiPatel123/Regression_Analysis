```python
#installing important library
import pandas as pd
import time
import glob    
import os
import numpy as np
```

```python
# 1. Define the path where your CSVs are stored
file_path = 'D:\Personal\project\Canada_Food_Regression\Data\descrete.xlsx'

# 2. Read all sheets at once by setting sheet_name=None
# This returns a dictionary: { "SheetName": DataFrame }
all_sheets_dict = pd.read_excel(file_path, sheet_name=None)

# 3. Combine them into one DataFrame
# We'll add a 'Sheet_Origin' column so you know which year/month the data belongs to
combined_df = pd.concat(
    [df.assign(Sheet_Origin=name) for name, df in all_sheets_dict.items()], 
    ignore_index=True
)

# 4. Preview the combined data
print(f"Successfully combined {len(all_sheets_dict)} sheets.")
print(combined_df.head())

# 5. Save to a true CSV for your model
combined_df.to_csv('D:\Personal\project\Canada_Food_Regression\Data\combined_data.csv', index=False)
```
