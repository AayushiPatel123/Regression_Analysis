```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# 1. Load the dataset
df = pd.read_csv('D:\Personal\project\Canada_Food_Regression\Data\combined_data.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
# Identify top 5 regions based on latest available data
latest_date = df['Date'].max()
top_provinces = df[df['Date'] == latest_date].sort_values('Composite_HPI_SA', ascending=False).head(5)['Province'].tolist()
```

```python
# Define future dates once for all plots
future_dates = [latest_date + pd.DateOffset(months=i) for i in range(1, 25)]
```

```python
# --- THE LOOP ---
for i, region in enumerate(top_provinces):
    # CRITICAL: Create a NEW figure for every single province
    plt.figure(figsize=(12, 6)) 
    
    # Filter and Prepare
    df_p = df[df['Province'] == region].sort_values('Date').copy()
    df_p['Time_Index'] = np.arange(len(df_p))
    
    # Model & Flatten Predictions
    X = df_p[['Time_Index']]
    y = df_p['Composite_HPI_SA']
    model = LinearRegression().fit(X, y)
    y_pred_hist = model.predict(X).flatten()  # Flatten fixed previous error
    std_error = np.std(y - y_pred_hist)
    
    # Forecast 2026
    future_idx = np.array(range(len(df_p), len(df_p) + 24)).reshape(-1, 1)
    preds = model.predict(future_idx).flatten()  # Flatten fixed previous error
    
    # Plotting
    plt.scatter(df_p['Date'], y, color='blue', alpha=0.2, s=8, label='Actual Data')
    plt.plot(df_p['Date'], y_pred_hist, color='red', label='Trend Line')
    plt.plot(future_dates, preds, color='green', linestyle='--', label='2026 Forecast')
    plt.fill_between(future_dates, preds - 1.96*std_error, preds + 1.96*std_error, color='green', alpha=0.1, label='95% CI')
    
    # Aesthetics
    plt.title(f'Housing Market Outlook: {region}', fontsize=14)
    plt.ylabel('Composite House Price Index')
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # --- SAVE WITH UNIQUE NAME ---
    file_name = f"forecast_{i+1}_{region.lower().replace(' ', '_')}.png"
    plt.savefig(f"D:\\Personal\\project\\Canada_Food_Regression\\.vscode\\Figures\\{file_name}")
    print(f"Successfully generated graph {i+1}: {file_name}")
    
    # CRITICAL: Close the figure so the next loop starts a fresh one
    plt.close() 

print("Done! Check your folder for all 5 PNG files.")
```
